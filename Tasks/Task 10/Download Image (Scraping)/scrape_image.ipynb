# Imports

```python
import requests
from bs4 import BeautifulSoup
```

# Extracting Web page content

```python
page = requests.get("https://en.wikipedia.org/wiki/Albert_Einstein").text
soup = BeautifulSoup(page,"lxml")
print(soup.prettify())
```

# Extracting info Container

```python
info_table = soup.find('table',class_ = "infobox biography vcard")
print(info_table.prettify())
```

# Extracting Image link

```python
image_link = "https:" + info_table.tbody.find_all('tr')[1].td.span.a.img.attrs['src']
image_link
```

# Downloading & Saving image

```python
image = requests.get(image_link)
with open("Einstein.png","wb") as file:
    file.write(image.content)
```
