# Imports

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

# Launching Website using selenium

```python
driver = webdriver.Chrome()
driver.get("https://www.youtube.com/results?search_query=python&sp=EgIYAQ%253D%253D")
content = driver.page_source.encode('utf-8').strip()
```

# Extracting Web page code

```python
soup = BeautifulSoup(content, "lxml")
```

```python
videos_container = soup.find_all('div',class_="style-scope ytd-video-renderer")
print (videos_container)
```

```python
print(videos_container[10].prettify())
```

# Creating Video Container Model

```python
class VideoModel:
    def __init__(self,title,link):
        self.title = title
        self.link = link
```

# Extracting Video title and link

```python
videos_models=[]

for video in videos_container:
    if(video.find('yt-formatted-string',class_="style-scope ytd-video-renderer")) :
        title = video.find('yt-formatted-string',class_="style-scope ytd-video-renderer")
    else :
        continue
    if(video.find('a',{"id": "thumbnail", "class" : "yt-simple-endpoint inline-block style-scope ytd-thumbnail"})):
        if(video.find('a',{"id": "thumbnail", "class" : "yt-simple-endpoint inline-block style-scope ytd-thumbnail"}).has_attr('href')) :
            link = video.find('a',{"id": "thumbnail", "class" : "yt-simple-endpoint inline-block style-scope ytd-thumbnail"}).attrs['href']
        else :
            continue
    else :
        continue  
    video_model = VideoModel(title= title.get_text(), link= "https://www.youtube.com/" + link)
    videos_models.append(video_model)

```

```python
for video_model in videos_models:
    print(video_model.title)
    print(video_model.link)
    print("--------------------------------")
print(len(videos_models))
```

# Downloaing Video

```python
from pytube import YouTube

def Download(link,resolution):
    yt = YouTube(link)
    yt.streams.filter(res=resolution,file_extension='mp4')
    stream = yt.streams.get_by_itag(22) 
    stream.download()
```

```python
Download(link=videos_models[1].link,resolution="720p")
```
